```python
import pandas as pd
pd.options.display.max_columns = None # Muestra todas las columnas de los dataframes
import json
import warnings
warnings.filterwarnings("ignore") # No muestra advertencias 
import re
import ast
import pyarrow

# Leo el dataframe auxiliar f5aux
df = pd.read_parquet("Tablas/f5aux.parquet")
df
```

```python
# Creo columnas con los 3 valores diferentes de sentiment
df_pivot = df.pivot_table(index="developer", columns="sentiment", values="item_id", aggfunc="count")
df_pivot.head(10)
```

```python
# Reemplazo nulos con ceros
a = df_pivot.fillna(0)
a
```

```python
# Transformo todos los números a enteros
a = a.astype(int)
a
```

```python
# Veo como funciona la consulta
developer = 'tobyfox'
# Busco los valores de sentiment para el developer ingresado
a0 = a.loc[developer, 0]
a1 = a.loc[developer, 1]
a2 = a.loc[developer, 2]
# Muestro la respuesta en el formato requerido
print(f"{{\'{developer}\':[Negative={a0}, Neutral={a1}, Positive={a2}]}}")

```
