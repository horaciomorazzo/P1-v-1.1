```python
import pandas as pd
import json
import a
import re
import ast


import warnings
warnings.filterwarnings("ignore")
```

```python
archivo = 'output_steam_games.json'
# Ruta al dataset australian_user_reviews


# Se lee de cada lÃ­nea del dataset
fs = []
with open(archivo, encoding='utf-8') as f:
    for line in f.readlines():
        data = json.loads(line)
        fs.append(data)

# Se convierte en dataframe
df_1 = pd.DataFrame(fs)
df_1

```

```python
# Eliminamos las filas con todos los registros nulos
df_2 = df_1.dropna(how='all').reset_index(drop=True)
df_2.shape
```

```python
# Lo guardo en CSV para analizarlo visualmente
df_2.to_csv('steam_games_output.csv', index=False)
df_2
```

Analizando el CSV veo que hay juegos donde no existen datos del developer. No obstante los mantengo pues probablemente pueda necesitarlos para las recomendaciones.

```python
# Aplico una funciÃ³n a la columna release_date para quedarme con el aÃ±o de las fechas vÃ¡lidas (YYYY-MM-DD). 
# Crea columna nueva con el aÃ±o
df_2['release_year'] = df_2['release_date'].apply(a.obtener_anio_release)
# elimina la columna 'release_date'
df_3= df_2.drop('release_date', axis=1)
df_3.head(1)
```

```python
# Elimino las filas que no me van a servir.
df_4 = df_3.drop(['publisher', 'app_name', 'url', 'tags',  'reviews_url', 'specs', 'price', 'early_access'], axis = 1)
```

```python
# Ahora elimino duplicados en id
df_4.drop_duplicates(subset='id', inplace=True, keep='first')
df_4.reset_index(drop=True, inplace=True) 
df_4
```

Los nombres de los juegos realizan una mezcla intencionada de mayÃºsculas y minÃºsculas, o palabras sin sentido aparente. BuscÃ¡ndolos en **https://store.steampowered.com/** compruebo que efectivamente estÃ¡n bien escritos en el dataframe, con lo cual no realizarÃ© ningÃºn tipo de tratamiento a dichas cadenas de texto.

```python
df_4
```

```python
nulos = df_4["developer"].isna().sum()
print('Nulos developer ', nulos)
nulos = df_4["title"].isna().sum()
print('Nulos title ', nulos)
```

```python
# Eliminamos los nulos en base a title. Existen tÃ­tulos donde no hay informaciÃ³n del desarrollador pero aÃºn asÃ­ pueden ser de utilidad en las recomendaciones, por ello los mantenemos.
df_5 = df_4.dropna(subset=['title'])
```

```python
df_5
```

Elimino las filas con el id corrompido. Son aquellas con el string '&amp' en la columna title Ã³ genres.

```python
df_5 = df_5.explode('genres')
df_5 = df_5.dropna(subset=['genres'])
df_5.head()
```

```python
# Eliminamos filas que contengan '&amp' y 'Dato no disponible'
condicion = df_5['genres'].str.contains('&amp', regex=True) | df_5['title'].str.contains('&amp', regex=True) | df_5['developer'].str.contains('&amp', regex=True) | df_5['release_year'].str.contains('Dato no disponible', regex=True)
df_6 = df_5[~condicion]
```

```python
df_6.reset_index(drop=True, inplace=True)
df_6
```

```python
# Elimino todas la filas que tienen valores nulos en 'id' o 'developer'
df_7 = df_6.dropna(subset=['id', 'developer'])
# Reconstruyo Ã­ndices
df_7.reset_index(drop=True, inplace=True)
df_7
```

```python
# Convierto el tipo de datos de las columnas id y release_year a entero
df_7['release_year'] = df_7['release_year'].astype(int)
df_7['id'] = df_7['id'].astype(int)

```

```python
# Guardo el dataset con formato CSV
archivo = 'steam_games_output_limpio.csv'
df_6.to_csv(archivo, index=False)
# Lo guardo en parquet para la API
df_6.to_parquet('steam_games_output_limpio.parquet', compression='gzip')

```
