```python
import pandas as pd
import ast

import warnings
warnings.filterwarnings("ignore")
```

```python
#Cargamos JSON
archivo = 'australian_users_items.json'

# Lo leemos línea a línea
fs = []
with open(archivo, encoding = 'utf-8') as file:
    for line in file.readlines():
        fs.append(ast.literal_eval(line))

# Armamos el dataframe
df_1 = pd.DataFrame(fs)
df_1
```

```python
# Eliminamos las filas con todos los registros nulos (si las hubiere)
df_2 = df_1.dropna(how='all').reset_index(drop=True)
df_2.shape
```

```python
# Elimino duplicados
df_2 = df_2.drop_duplicates(subset="user_id")
```

```python
df_2.shape
```

```python
# Elimino las filas donde items_count = 0
df_2 = df_2.drop(df_2[df_2["items_count"] == 0].index)
df_2.reset_index(drop=True, inplace=True)
```

```python
df_2
```

```python
# Como vimos en el shape no había registros nulos. Ahora procedemos a desanidar y quedarnos únicamente con las columnas que nos serán de utilidad más adelante. 
item_e = []
for i, row in df_2.iterrows():
    user_id = row['user_id']
    
    for j in row['items']:
        item_1= {
                'user_id': user_id,        
                'item_id': j['item_id'],
                'item_name': j['item_name'],
                'playtime_forever': j['playtime_forever']
                }
        item_e.append(item_1) # Agremos a desanidado el diccionario item_new
df_3= pd.DataFrame(item_e)
```

```python
df_3
```

```python
df_3.drop(df_3[df_3["playtime_forever"] == 0].index, inplace=True)
```

```python
df_3
```

```python
df_3.dtypes
```

```python
# Convertimos los registros de la columna 'item_id' a enteros de 64 bit
df_3['item_id'] = df_3['item_id'].astype('int64')
# Convertimos los registros de la columna 'user_id' a string
df_3['user_id'] = df_3['user_id'].astype(str)

```

```python
# Lo guardo en csv 
archivo = 'user_items_limpio.csv'
df_3.to_csv(archivo, index=False)
# Lo guardo en parquet para ser utilizado en la api
df_3.to_parquet('user_items_limpio.parquet', compression='gzip')
```
